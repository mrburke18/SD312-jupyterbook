```python
import numpy as np
import sklearn.decomposition
```

```python
N=50
M=5
k=2

A=np.random.normal(size=(N,k))@np.random.normal(size=(k,M))
```

```python
A=A-np.min(A)
A=(5/np.max(A))*A
```

```python
A=np.round(A)
A
```

```python
pca=sklearn.decomposition.PCA(2)
pca.fit(A)
pca.components_[0]
```

```python
import plotly.express as px
```

```python
fig=px.bar(x=['Red','Green','Blue','Purple','Black'],y=pca.components_[0])
          #paper_bgcolor="white",  # Set background to white
          #plot_bgcolor="white",  # Set plot area to white
          #margin=dict(l=0, r=0, t=0, b=0),  # Remove margins
          #showlegend=False,  # Remove legend (optional)

fig.update_traces(marker=dict(color="gray", line=dict(color="black")))
fig.show()
```
