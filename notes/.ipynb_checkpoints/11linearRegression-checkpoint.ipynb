# Linear Regression

These notes are a complement to your reading.  For the purposes of these notes, $X$ is the $n\times k$ data matrix of all your data points, including a column for the bias.  $x_i$ refers to a single row of that matrix, that is, one data point.  $y_i$ is a single data point's target value, and $y$ is the vector of all target values.  This means that when performing linear regression we are trying to find the $k\times 1$ vector $w$ such that $Xw\approx y$.

First, why are we using $\approx$, and not =?  Why can we not find a $w$ which makes this exactly right?  Remember that *noise* can come from a variety of sources.  Measurement error, mislabelling, etc. can all result in an incorrect $y$.  So can the fact that world just isn't deterministic - there's randomness, and so $y_i$ can be different if measured twice for the same $x_i$.  Finally, it may be that you've chosen the wrong measurements to make for your vector $x_i$, and it just isn't fully predictive of the phenomenon you're measuring.

In linear algebra, this means that the vector $y$ is not in the vector space spanned by $X$, which means that no $w$ exists which makes $Xw=y$ into an equality.  So, we're stuck with finding the $w$ which makes it as close as possible.

In your reading, this was described as an attempt to find the $w$ which minimizes the sum of squares error $\sum_i (x_iw-y_i)^2$.  By taking the gradient of that term, setting it equal to zero (as with minimizing functions in calculus class), and solving for $w$, we arrived at $w=(X^TX)^{-1}X^Ty$.  There are a number of other ways of looking at this problem.  The first is...

### Algebraically

OK, so you have $Xw\approx y$, and you'd like to solve for $w$.  In regular algebra, this would be easy: you'd divide both sides by $X$, giving you $w$.  Unfortunately, you can't divide in linear algebra, you can only multiply by the inverse.  However, $X$ is probably not square, and so does not have an inverse.  So, what to do?  Well, the first step is to MAKE it square, by left-multiplying by its transpose: $X^TXw\approx X^Ty$.

$X^TX$ is $k\times k$, and so is square.  Now, this doesn't necessarily mean
that $X^TX$ is invertible, it has to be full rank for that.  Without diving too deep into linear algebra (too late?), if $X$ is full rank, then $X^TX$ is, too.  Let's assume that condition is met.

So, we can now left-multiply by $(X^TX)^{-1}$, which we just decided probably exists, giving us $(X^TX)^{-1}X^TXw\approx (X^TX)^{-1}X^Ty$.  A matrix and its inverse cancel out, leaving us with $w\approx (X^TX)^{-1}X^Ty$.  And there we are.

For this reason, $(X^TX)^{-1}X^T$ has a number of special names.  First, it's known as the *Moore-Penrose pseudo-inverse* of $X$, and it is often useful when you don't have an inverse, but need to "divide."  Second, it finds the point in the span of $X$ which is closest to $y$: it's the *projection* of $y$ onto $X$.  For this reason, it's also known as the *projection operator*.  

Note that even if an inverse exists, it's often computationally more stable to compute the pseudoinverse instead.  The most common algorithm for calculating this pseudoinverse actually involves... the magical SVD.  You can use this algorithm most simply in numpy using the operation [`np.linalg.pinv`](https://docs.scipy.org/doc/numpy-1.15.0/reference/generated/numpy.linalg.pinv.html).

### Probablistically

Let's pretend that a linear combination of our features of $X$ *are* capable of fully explaining $y$, it's just that the world has thrown into the mix some random Gaussian error, so that $y_i=x_iw+\epsilon_i$, where $\epsilon_i$ is drawn from the normal distribution (aka the Gaussian, aka the bell curve) centered around 0.  This means $\epsilon_i$ is close to zero, but is probably not zero.  It also means that given a $x_i$ and a $w$, we can describe each $y_i$ as likely (because the associated $\epsilon_i$ is small, and so likely to come from the Gaussian), or unlikely (because $\epsilon_i$ is not close to zero).  In fact, we can write down $p(y|X,w)$, and then solve for the $w$ which makes that probability as large as possible.  This is known as the *maximum likelihood* solution.

The result?  $w=(X^TX)^{-1}X^Ty$.



## An Example

This demo is intended to show how feature transformations can be useful in building better linear regression models.

This notebook uses several helper functions for generating data, performing linear regression, and plotting the results. Here is a breakdown of their purposes:
- `genData1d(N, noise=1)`: Generates `N` data points in 2D space that roughly follow a line. It creates a random line and then adds Gaussian noise to the y-values.
- `genData2d(N, noise=1)`: Generates `N` data points in 3D space that roughly lie on a plane. It creates a random plane and adds Gaussian noise to the z-values.
- `genData(N, noise=1)`: Generates `N` data points in 2D that follow a parabolic (quadratic) curve, with added noise.
- `linReg(data, ys)`: Performs linear regression using the pseudo-inverse method to find the optimal weights w that minimize the squared error between the predicted and actual y-values.
- `plotPoints(xs, ys)`: Creates a simple 2D scatter plot of the data.
- `plotItLin(xs, ys, w)`: Visualizes the results of 1D linear regression by plotting the original data points and the line defined by w.
- `plot3D(data, ys, w=None)`: Creates a 3D scatter plot of the data. If the weight vector w is provided, it also plots the plane defined by w.

```python
import numpy as np
import matplotlib.pyplot as plt
import argparse
from mpl_toolkits.mplot3d import Axes3D
import plotly.express as px
import plotly.graph_objects as go

np.random.seed(0)
```

```python
'''
Generate N 2d datapoints, plus noise
'''
def genData1d(N,noise=1):
  line=np.random.randn(2) #random A,B for Ax+B=y
  xs=np.random.random(N)*20-10 #random points on x-axis between -10 and 10

  #for each of those points, calculate x,and 1
  data=np.asarray( [[x,1] for x in xs] )
  ys=data@line #calculate resulting y value
  
  #for each, add some random error
  ys=ys+noise*np.random.randn(N)

  return xs,ys

'''
Generate N 3d datapoints, plus noise
'''
def genData2d(N,noise=1):
  line=np.random.randn(3) #random A,B for Ax_1+Bx_2+C=y
  xs=np.random.random((N,2))*20-10 #random points (-10,-10) and (10,10)

  #for each of those points, calculate x,and 1
  data=np.asarray( [[x[0],x[1],1] for x in xs] )
  ys=data@line #calculate resulting y value
  
  #for each, add some random error
  ys=ys+noise*np.random.randn(N)

  return xs,ys

'''
Draw the line of best fit in x,y space.
'''
def plotItLin(xs,ys,w):
    fig=px.scatter(x=xs,y=ys)
    line=go.Scatter(x=[-10,10],y=np.array([[-10,1],[10,1]])@w)
    fig.add_trace(line)
    fig.update_layout(
        xaxis_title="x",
        yaxis_title="y",
        height=600,
        width=800
    )
    fig.show()

'''
Fancy-shmancy drawing of points and plane of best fit in x,x^2 space
'''
def plot3D(data,ys,w=None):
    if w is None:
        fig=px.scatter_3d(x=data[:,0],y=data[:,1],z=ys)
    else:
        scat=go.Scatter3d(x=data[:,0],y=data[:,1],z=ys,mode='markers')
        X1=np.arange(data[:,0].min(),data[:,0].max())
        X2=np.arange(data[:,1].min(),data[:,1].max())
        plane=np.asarray([[x1,x2,1] for x2 in X2 for x1 in X1])@w
        X1,X2=np.meshgrid(X1,X2)
        plane_trace=go.Mesh3d(
            x=X1.flatten(),
            y=X2.flatten(),
            z=plane,
            opacity=.15,
            color='blue',
            showscale=False
        )
        fig=go.Figure(data=[scat,plane_trace])
            
    fig.update_layout(scene=dict(xaxis_title='x1',yaxis_title='x2',zaxis_title='label'),
                     height=800,width=800)
    fig.update_scenes(xaxis_range=[data[:,0].min()-1,data[:,0].max()+1],yaxis_range=[data[:,1].min()-1,data[:,1].max()+1],zaxis_range=[ys.min()-1,ys.max()+1])
    fig.show()

def plotPoints(xs,ys):
    fig=px.scatter(x=xs,y=ys)
    fig.update_layout(
        xaxis_title="x",
        yaxis_title="y",
        height=600,
        width=800
    )
    fig.show()

'''
Do linear regression
'''
def linReg(data,ys):
    w=np.linalg.pinv(data)@ys
    return w
```

Let's generate some data that roughly lies on a line, and plot them.

```python
N=20
noise=.7

xs,ys=genData1d(N,noise=noise)
plotPoints(xs,ys)
```

Now we add a bias feature (note that the first column of `data` is just `xs`, and the second column is just ones), perform linear regression, and plot the line on top of the same points. Not bad!

```python
data=np.ones((N,2))
data[:,0]=xs
w=linReg(data,ys)
plotItLin(xs,ys,w)
```

Looks pretty good!

Now we're going to do the same thing where our points lie in two dimensions. Each datapoint now has a $x1$ and a $x2$ dimension, so we can build a data matrix like this (including the bias vector):

$$X=\left[\begin{array}{ccc}
x1_1&x2_1&1\\
x1_2&x2_2&1\\
\vdots&\\
x1_n&x2_n&1
\end{array}\right]$$

Note that the resulting `w` from linear regression now defines a plane which fits the data.

```python
xs,ys=genData2d(N,noise=noise)
data=np.ones((N,3))
data[:,:2]=xs
plot3D(data,ys)
w=linReg(data,ys)
plot3D(data,ys,w)
```

## Quadratic Plotting

OK, now we're going to generate some data that does not lie on a line and see how we can fit it.

```python
import pandas as pd
```

```python
def genData(N,noise=1):
    line=np.random.randn(3) #random A,B,C for Ax^2+Bx+C=0
    xs=np.random.random(N)*20-10 #random points on x-axis between -10 and 10

    #for each of those points, calculate x^2,x,and 1, and add in some random
    #error
    data=np.asarray( [[x*x,x,1+noise*np.random.randn()] for x in xs] )
    ys=data@line #calculate resulting y value
    return xs,ys
```

```python
N=20
noise=8
```

```python
xs,ys=genData(N,noise)
plotPoints(xs,ys)
```

You'd like to make a prediction, for other values of $x$ as to what the $y$ will be.  So, you dutifully do linear regression.  You build yourself a data matrix (taking into account the bias):

$$X=\left[\begin{array}{cc}
x_1&1\\
x_2&1\\
\vdots&\\
x_n&1
\end{array}\right]$$

and solve for the 2-element vector $w$, which gives you a line $y_{pred}=mx+b$, which looks like this:

```python
data = np.ones((N,2))
data[:,0]=xs
w=linReg(data,ys)
```

```python
plotItLin(xs,ys,w)
```

Well that didn't work well at all. In fact, you note that this doesn't seem to be linear at all, it seems to be quadratic. So, you must be out of luck, since this is *linear* regression.

Of course, that's wrong. You have an epiphany, and build a new data matrix:

$$X=\left[\begin{array}{ccc}
x_1^2&x_1&1\\
x_2^2&x_2&1\\
\vdots&\vdots&\vdots\\
x_n^2&x_n&1
\end{array}\right]$$

Now, you have artificially created a third dimension, the $x^2$ dimension.  Here, your points look like this:

```python
data = np.ones((N,3))
data[:,0] = xs*xs
data[:,1] = xs
```

```python
plot3D(data,ys)
```

We can solve for a $w$ and the plane of best fit, just like in the example above.

```python
w=linReg(data,ys)
plot3D(data,ys,w)
```

That plane now fits pretty well! The three elements of the weight vector correspond to $A$, $B$, and $C$ in the function $Ax^2+Bx+C$, so we can take those same weights and plot that plane as a parabola in the original space.

```python
# Create a scatter plot of the original data points
fig = go.Figure(data=go.Scatter(x=xs, y=ys, mode='markers', name='Data Points'))

# Generate a smooth set of x-values for plotting the curve
x_curve = np.linspace(xs.min(), xs.max(), 100)

# Create the corresponding data matrix with x^2, x, and 1
data_curve = np.ones((100, 3))
data_curve[:, 0] = x_curve * x_curve
data_curve[:, 1] = x_curve

# Calculate the y-values for the parabola using the learned weights 'w'
y_curve = data_curve @ w

# Add the parabolic curve to the plot
fig.add_trace(go.Scatter(x=x_curve, y=y_curve, mode='lines', name='Parabolic Fit'))

# Update the layout for clarity
fig.update_layout(
    xaxis_title="x",
    yaxis_title="y",
    height=600,
    width=800
)

fig.show()
```

That parabola fits pretty well! It's a linear solution (a plane) in the 3d space, but you can draw it as a quadratic line in 2d space.

## More must be better!

OK, so adding another feature made things better. What if we add another, and another, and another? Let's add $x^3, x^4$, etc.

```python
N=12
noise=8
xs,ys=genData(N,noise)
```

```python
# Import the make_subplots function from plotly
from plotly.subplots import make_subplots

# --- Pre-calculate errors and create titles ---
def make_polynomial_subplots(xs,ys):
    subplot_titles = []
    for i in range(1, 10):
        # Create the data matrix with polynomial features
        data_poly = np.asarray([[x**p for p in range(i, -1, -1)] for x in xs])
        # Perform linear regression
        w = linReg(data_poly, ys)
        # Calculate predictions on training data to find the error
        y_pred = data_poly @ w
        # Calculate Mean Squared Error (MSE)
        mse = np.mean((ys - y_pred)**2)
        # Append the formatted title including the MSE
        subplot_titles.append(f'Degree {i}<br>MSE: {mse:.2f}')
    
    
    # Create a 3x3 subplot figure with the calculated titles.
    fig = make_subplots(rows=3, cols=3, subplot_titles=subplot_titles)
    
    # Generate a smooth range of x-values for plotting the fitted curves
    x_curve = np.linspace(xs.min(), xs.max(), 100)
    
    # Loop through degrees 1 to 9 again to perform the plotting
    for i in range(1, 10):
        # --- 1. Feature Engineering ---
        # Recreate the data matrix for the current degree
        data_poly = np.asarray([[x**p for p in range(i, -1, -1)] for x in xs])
    
        # --- 2. Linear Regression ---
        # Re-calculate the weights for the current degree
        w = linReg(data_poly, ys)
    
        # --- 3. Plotting ---
        # Determine the correct row and column for the current subplot
        row = (i - 1) // 3 + 1
        col = (i - 1) % 3 + 1
    
        # Add the original data points as a scatter plot
        fig.add_trace(
            go.Scatter(x=xs, y=ys, mode='markers', name='Data', showlegend=False),
            row=row, col=col
        )
    
        # Create the feature matrix for the smooth curve using the same polynomial degrees
        data_curve = np.asarray([[x**p for p in range(i, -1, -1)] for x in x_curve])
        
        # Calculate the y-values for the fitted curve using the learned weights 'w'
        y_curve = data_curve @ w
    
        # Add the polynomial curve as a line plot
        fig.add_trace(
            go.Scatter(x=x_curve, y=y_curve, mode='lines', name=f'Degree {i} Fit', showlegend=False),
            row=row, col=col
        )
    
    # Update the overall layout for a better presentation
    fig.update_layout(
        title_text="Effect of Increasing Polynomial Degree on Model Fit",
        height=800,
        width=900
    )
    
    fig.show()
    
make_polynomial_subplots(xs,ys)
```

We can see that the training error is indeed dropping, and it is doing a mathematically better job of fitting the datapoints. However, in order to do this, it's having to do more and more dramatic things with the line, such that it qualitatively doesn't fit the overall trend very well anymore. If we had a test set, error on that test set would be increasing!

Remember our Hoeffding inequality?  $P[|E_{in}(g)-E_{out}(g)|>\epsilon]\leq 2Me^{-2\epsilon^2N}$?  The complexity ($M$) has gotten larger, and so the gap between our training and testing error has increased.

Of course, we've learned that our best defence against this overfitting is to add more data. What if I did this same experiment with 20 datapoints?

```python
N=20
noise=8
xs,ys=genData(N,noise)
make_polynomial_subplots(xs,ys)
```

Somewhat less stupid (even though training error is higher)? Maybe? What about with 50 datapoints?

```python
N=50
noise=8
xs,ys=genData(N,noise)
make_polynomial_subplots(xs,ys)
```

Not stupid at all. The more data we have, the more powerful a model we can get away with.
