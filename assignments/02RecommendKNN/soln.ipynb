# k-NN Recommendation systems

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
import csv
```

```python
dataset='ml-25m'
```

```python
ratings=pd.read_csv(dataset+'/ratings.csv')
max(ratings.userId),max(ratings.movieId)
USERS=max(ratings.userId)
MOVIES=max(ratings.movieId)
A=np.zeros((USERS,MOVIES))
for user,movie,rate in zip(ratings.userId,ratings.movieId,ratings.rating):
    A[user-1,movie-1]=rate #I prefer indexing from 0
```

```python
movies=dict() #make a dictionary that goes from movieID-1 to the movie name
with open(dataset+'/movies.csv') as f:
    f.readline()
    reader=csv.reader(f,delimiter=',')
    for row in reader:
        key=int(row[0])-1
        movies[key]=row[1]
```

```python
def makeNN(metric='minkowski'):
    nn=NearestNeighbors(n_neighbors=11,metric=metric)
    nn.fit(A.T)
    return nn
```

```python
nn=NearestNeighbors(n_neighbors=11)
nn.fit(A.T)
```

```python
nn.kneighbors(A.T[18,:].reshape(1,-1))
```

```python
def getNN(movieID):
    dist,ids=nn.kneighbors(A.T[movieID-1].reshape(1,-1))
    namelist=[]
    for neigh in ids[0]:
        namelist.append(movies[neigh])
    return namelist
```
