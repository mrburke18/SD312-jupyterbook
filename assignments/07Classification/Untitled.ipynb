```python
import pandas as pd
```

```python
df=pd.read_csv('analytic_data2023.csv')
```

```python
df[['Premature Death raw value','Premature Death numerator','Premature Death denominator','Premature Death CI low','Premature Death CI high']].head()
```

```python
cols=list(df.columns[:5])
for col in df.columns:
    print(col)
    if 'raw value' in col or 'Ratio' in col:
        cols.append(col)
cols
```

```python
cutdown=df[df['County FIPS Code']!=0][cols]
```

```python
cutdown.head()
```

```python
elect=pd.read_csv('countypres_2000-2020.csv')
elect=elect[elect['year']==2020]
elect.head()
```

```python
elect=elect[['state','county_name','county_fips','candidate','party','candidatevotes','totalvotes']]
```

```python
elect.head()
```

```python
elect['vote_ratio']=elect['candidatevotes']/elect['totalvotes']
```

```python
sum(elect.candidate=='JO JORGENSEN')
```

```python
elect.head()
```

```python
len(elect.county_fips.unique())
```

```python
len(cutdown['5-digit FIPS Code'].unique())
```

```python
pivot_elect=elect.pivot_table(index='county_fips',columns='candidate',values='vote_ratio')
```

```python
pivot_elect.head()
```

```python
merged=pd.merge(cutdown,pivot_elect,left_on='5-digit FIPS Code',right_index=True,how='inner')
```

```python
merged['JO JORGENSEN'].fillna(0,inplace=True)
merged['OTHER'].fillna(0,inplace=True)
```

```python
merged
```

```python
nancounts=merged.isna().sum()
```

```python
merged.drop('Living Wage raw value',axis=1,inplace=True)
merged.drop('Children Eligible for Free or Reduced Price Lunch raw value',axis=1,inplace=True)
```

```python
merged.drop('Residential Segregation - Black/White raw value',axis=1,inplace=True)
```

```python
merged.drop('Child Care Cost Burden raw value',axis=1,inplace=True)
```

```python
merged.drop('Child Care Centers raw value',axis=1,inplace=True)
```

```python
for col,val in zip(merged.columns,nancounts):
    print(col,val)
```
