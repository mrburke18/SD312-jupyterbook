```python
import numpy as np
```

```python
ratings=np.array([[5,2,4.5],[1,4,1],[4,1,4],[2,2,1.5]])
print(ratings)
```

```python
u,s,vt=np.linalg.svd(ratings)
```

```python
print(u.shape)
print(u)
print()
print(s.shape)
print(s)
print()
print(vt.shape)
print(vt)
```

```python
Sigma=np.zeros((4,3))
Sigma[:3,:3]=np.diag(s)
print(Sigma)
```

```python
print(u@Sigma@vt)
```
