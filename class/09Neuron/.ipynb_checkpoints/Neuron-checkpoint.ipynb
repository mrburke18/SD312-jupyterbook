```python
import numpy as np
import plotly.graph_objects as go
```

```python
array1=np.linspace(-1,1)
array2=np.linspace(-1,1)
data = np.array(np.meshgrid(array1, array2)).T.reshape(-1, 2)
print(data.shape)
labels=data[:,1]>data[:,0]*data[:,0]
labels=labels.astype(int)
print(labels.shape)

scatter=go.Scatter(x=data[:,0],y=data[:,1],mode='markers',marker=dict(size=3,color=labels))
layout=go.Layout(height=500)
fig=go.Figure(data=[scatter],layout=layout)
fig.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
class MyNeuron(nn.Module):
    def __init__(self):
        super(MyNeuron, self).__init__()
        self.layer = nn.Linear(2, 1)
        self.activation = nn.Sigmoid()  # You can choose another activation function if needed

    def forward(self, x):
        x = self.layer(x)
        x = self.activation(x)
        return x
```

```python
mn=MyNeuron()

criterion=nn.MSELoss()
optimizer=optim.SGD(mn.parameters(), lr=.001)

xs=torch.tensor(data, dtype=torch.float32)
ys=torch.tensor(labels.reshape(-1,1), dtype=torch.float32)
    
EPOCHS=100000

for epoch in range(EPOCHS):
        
    predictions=mn(xs)
    loss=criterion(predictions, ys)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    if epoch%10000==0:
        print(f'epoch: {epoch} loss: {loss.item():.4f}')
```

```python
predictions=mn(xs).detach()
scatter=go.Scatter(x=data[:,0],y=data[:,1],mode='markers',marker=dict(size=3,color=predictions.round().numpy().flatten()))
layout=go.Layout(height=500)
fig=go.Figure(data=[scatter],layout=layout)
fig.show()
```

```python
newdata=data.copy()
newdata[:,0]=newdata[:,0]*newdata[:,0]

mn=MyNeuron()

criterion=nn.MSELoss()
optimizer=optim.SGD(mn.parameters(), lr=.001)

xs=torch.tensor(newdata, dtype=torch.float32)
ys=torch.tensor(labels.reshape(-1,1), dtype=torch.float32)
    
EPOCHS=100000

for epoch in range(EPOCHS):
        
    predictions=mn(xs)
    loss=criterion(predictions, ys)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    if epoch%10000==0:
        print(f'epoch: {epoch} loss: {loss.item():.4f}')
```

```python
predictions=mn(xs).detach()
scatter=go.Scatter(x=data[:,0],y=data[:,1],mode='markers',marker=dict(size=3,color=predictions.round().numpy().flatten()))
layout=go.Layout(height=500)
fig=go.Figure(data=[scatter],layout=layout)
fig.show()
```

```python
# Define the neural network class with a single hidden layer
class LearnInputs(nn.Module):
    def __init__(self):
        super(LearnInputs, self).__init__()
        self.hidden_layer = nn.Linear(2, 2)
        self.activation = nn.Sigmoid()
        self.output_layer = nn.Linear(2, 1)
        self.output_act = nn.Sigmoid()

    def forward(self, x):
        x = self.hidden_layer(x)
        x = self.activation(x)
        x = self.output_layer(x)
        x = self.output_act(x)
        return x

li=LearnInputs()

criterion=nn.MSELoss()
optimizer=optim.SGD(li.parameters(), lr=.01)

xs=torch.tensor(data, dtype=torch.float32)
ys=torch.tensor(labels.reshape(-1,1), dtype=torch.float32)
    
EPOCHS=200000

for epoch in range(EPOCHS):
        
    predictions=li(xs)
    loss=criterion(predictions, ys)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    if epoch%10000==0:
        print(f'epoch: {epoch} loss: {loss.item():.4f}')
```

```python
predictions=li(xs).detach()
scatter=go.Scatter(x=data[:,0],y=data[:,1],mode='markers',marker=dict(size=3,color=predictions.round().numpy().flatten()))
layout=go.Layout(height=500)
fig=go.Figure(data=[scatter],layout=layout)
fig.show()
```
