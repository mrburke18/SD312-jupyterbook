```python
import numpy as np
```

```python
R=np.array([0,1,0,0])
V=np.array([0,0,0,0])
```

```python
for i in range(1000):
    Vp = np.array([V[1], V[2], V[1], V[2]])
    V = R+.9*Vp
print(V)
```

```python
import plotly.graph_objects as go

categories = ['s0','s1','s2','s3']
valueL = [0, 1, .9, .81]
valueR = [.9, 1, 0, 0]

left = go.Bar(x=categories, y=valueL, name='always left')
right = go.Bar(x=categories, y=valueR, name='always right')
both = go.Bar(x=categories, y=V, name='optimal')

fig = go.Figure(data=[left, right, both])

fig.update_layout(
    title='Value of policies',
    xaxis_title='States',
    yaxis_title='Value',
    plot_bgcolor='white',  # Setting plot background to white
    paper_bgcolor='white',  # Setting paper background to white
    font=dict(color='black')  # Setting font color to black
)

fig.write_image('leftValue.png',width=1000,height=500, scale=3)
```
